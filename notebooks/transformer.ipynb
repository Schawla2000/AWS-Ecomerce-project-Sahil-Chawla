Tranforming logic for Customer dataset

```python
class CustomerTransformer(BaseTransformer):
    def transform(self,df):
        return df.dropDuplicates(["customer_id"]).withColumn("email",lower(trim(col("email"))))
```

Tranforming logic for Order dataset

```python
class OrderTransformer(BaseTransformer):
    def transform(self, df):
        return df.filter(col("status") != "Cancelled")
```

Tranforming logic for Payment dataset

```python
class PaymentTransformer(BaseTransformer):
    def transform(self, df):
        return df.filter(col("status") == 'Paid')
```

Tranforming logic for Product dataset

```python
class ProductTransformer(BaseTransformer):
    def transform(self, df):
        return df.dropDuplicates(["product_id"])
```
